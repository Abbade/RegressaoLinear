```python
import pandas as pd
from sklearn.datasets import load_boston
```

```python
dataset = load_boston()
```

```python
dataset = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
dataset.head()
```

```python
print(dataset.shape)
```

```python
import seaborn as sns

%matplotlib inline
```

```python
sns.pairplot(dataset, x_vars=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'TAX', 'PTRATIO', 'B', 'LSTAT'], y_vars='LSTAT', size=7, aspect=0.7, kind='reg')
```

```python
x = dataset[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'TAX', 'PTRATIO', 'B', 'LSTAT']]
feature_cols = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'TAX', 'PTRATIO', 'B', 'LSTAT']
x.head()
```

```python
print(type(x))
print(x.shape)
```

```python
y = dataset['LSTAT']
y = dataset.LSTAT
y.head()
```

```python
print(type(y))
print(y.shape)
```

```python
from sklearn.cross_validation import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
```

```python
linreg.fit(x_train, y_train)
```

```python
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=1, normalize=False)
```

```python
print(linreg.intercept_)
print(linreg.coef_)
```

```python
list(zip(feature_cols, linreg.coef_))
```

```python
y_pred = linreg.predict(x_test)
```

```python
from sklearn import metrics
```

```python
import numpy as np
```

```python
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
